```
import numpy as np
from sklearn import datasets, svm

digits = datasets.load_digits()
images_and_labels = list(zip(digits.images, digits.target))
X = list()
for dig in images_and_labels:
  temp = list()
  for i in range(8):
    for j in range(8):
      temp.append(dig[0][i][j])
  X.append(temp)
X = np.array(X)
y = list()
for dig in images_and_labels:
  y.append(dig[1])

y = np.array(y)
y.reshape(1, -1)
clf = svm.SVC(kernel='rbf', C = 0.98)
clf.fit(X,y)
s = len(images_and_labels)
print("Incorrect:")
for i in range(len(images_and_labels)):
  if [images_and_labels[i][1]]!=clf.predict([X[i]]):
    s = s - 1
    print([images_and_labels[i][1]], clf.predict([X[i]]))
print("Accuracy:", s/len(images_and_labels))
```
